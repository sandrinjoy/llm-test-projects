```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of angles from 0 to 2*pi
theta = np.linspace(0, 2 * np.pi, 400)

# List of norms (p-values) and corresponding labels
norms = [1, 2, 3]
labels = ["L1 Norm", "L2 Norm", "L3 Norm"]

# Set up the plot
plt.figure(figsize=(6, 6))

# Calculate and plot the unit circle for each norm
for p, label in zip(norms, labels):
    # Compute the radial distance for each theta using the formula:
    # r = 1 / (|cos(theta)|^p + |sin(theta)|^p)^(1/p)
    r = 1 / (np.abs(np.cos(theta))**p + np.abs(np.sin(theta))**p)**(1/p)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    
    plt.plot(x, y, label=label)

# Draw x and y axes through the origin for clarity
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Set equal aspect ratio so the shapes are not distorted
plt.gca().set_aspect('equal')

# Add title and labels
plt.title("Unit 'Circles' for L1, L2, and L3 Norms")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_points(num_points=1000):
    # Generate angles
    theta = np.linspace(0, 2*np.pi, num_points)
    
    # L2 norm (regular circle)
    l2_x = np.cos(theta)
    l2_y = np.sin(theta)
    
    # L1 norm (diamond shape)
    l1_x = []
    l1_y = []
    for t in theta:
        if 0 <= t <= np.pi/2:  # First quadrant
            x = 1 / (1 + np.tan(t))
            y = np.tan(t) / (1 + np.tan(t))
        elif np.pi/2 < t <= np.pi:  # Second quadrant
            t_temp = np.pi - t
            x = -1 / (1 + np.tan(t_temp))
            y = np.tan(t_temp) / (1 + np.tan(t_temp))
        elif np.pi < t <= 3*np.pi/2:  # Third quadrant
            t_temp = t - np.pi
            x = -1 / (1 + np.tan(t_temp))
            y = -np.tan(t_temp) / (1 + np.tan(t_temp))
        else:  # Fourth quadrant
            t_temp = 2*np.pi - t
            x = 1 / (1 + np.tan(t_temp))
            y = -np.tan(t_temp) / (1 + np.tan(t_temp))
        l1_x.append(x)
        l1_y.append(y)
    
    # L3 norm
    r = 1 / (np.abs(np.cos(theta))**3 + np.abs(np.sin(theta))**3)**(1/3)
    l3_x = r * np.cos(theta)
    l3_y = r * np.sin(theta)
    
    return (l1_x, l1_y), (l2_x, l2_y), (l3_x, l3_y)

# Create the plot
plt.figure(figsize=(10, 10))

# Generate and plot points
(l1_x, l1_y), (l2_x, l2_y), (l3_x, l3_y) = generate_points()

# Plot each norm
plt.plot(l1_x, l1_y, 'b-', label='L1 norm', linewidth=2)
plt.plot(l2_x, l2_y, 'r-', label='L2 norm', linewidth=2)
plt.plot(l3_x, l3_y, 'g-', label='L3 norm', linewidth=2)

# Add grid and labels
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Unit Circles for Different Norms')
plt.legend()

# Set equal aspect ratio and axis limits
plt.axis('equal')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)


# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate angles from 0 to 2π
theta = np.linspace(0, 2 * np.pi, 1000)

# Calculate radii for each norm
r_l1 = 1 / (np.abs(np.cos(theta)) + np.abs(np.sin(theta)))
r_l2 = np.ones_like(theta)  # L2 norm (standard circle)
r_l3 = 1 / (np.abs(np.cos(theta))**3 + np.abs(np.sin(theta))**3)**(1/3)

# Convert polar to Cartesian coordinates
x_l1, y_l1 = r_l1 * np.cos(theta), r_l1 * np.sin(theta)
x_l2, y_l2 = r_l2 * np.cos(theta), r_l2 * np.sin(theta)
x_l3, y_l3 = r_l3 * np.cos(theta), r_l3 * np.sin(theta)

# Plotting
plt.figure(figsize=(8, 8))
plt.plot(x_l1, y_l1, label='L1 Norm (Diamond)')
plt.plot(x_l2, y_l2, label='L2 Norm (Circle)')
plt.plot(x_l3, y_l3, label='L3 Norm (Superellipse)')
plt.axis('equal')
plt.title('Unit Circles for L1, L2, and L3 Norms')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
